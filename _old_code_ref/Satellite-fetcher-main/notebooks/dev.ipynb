```python
import cdsapi
import pandas as pd
```

```python
dataset = "cams-global-reanalysis-eac4"
request ={'date': ['2024-08-01'], 'time': ['12:00'], 'data_format': 'netcdf_zip', 'variable': ['total_aerosol_optical_depth_550nm', 'total_column_water_vapour'], 'area': [28.86, -2.83, 30.89, -1.06]}

client = cdsapi.Client()
results = client.retrieve(dataset, request)
```

```python
results.download(target='CAMS_2024_12_01.nc')
```
